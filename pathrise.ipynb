### Features short description
**id** = unique id of a fellow </br>
**pathrise_status** = status of a fellow in the program: *Active .. Closed .. Withdrawn ... etc* </br>
**primary_track** = field of track: *Data .. Design .. Web .. etc* </br>
**cohort_tag** = Cohort tag: *monthname|year|(A/B) for start on first/third week of month* </br>
**program_duration_days** = how many days a fellow was in the program, N/A for current fellows </br>
**placed** = whether a fellow was placed while going through the program (pathrise_status==placed) </br>
**employment_status** = employment status when joined the program: *Employed .. copntractor .. etc* </br>
**highest_level_of_education** = highest level of education when joined the program   </br>
**length_of_job_search** = length of a job search before joining the program  </br>
**biggest_challenge_in_search** = biggest challenge in a job search (self-reported) </br>
**professional_experience** = professional work exeperience before joining the program  </br>
**work_authorization_status** = work authorization status before joining the program </br>
**number_of_interviews** = number of interviews before joining the program (self-reported) </br>
**number_of_applications** = number of applications before joining the program (self-reported) </br>
**gender** = gender </br>
**race** = race </br>
